```python
import pandas as pd
```

```python
df_weather= pd.read_csv('~/data/CSV_Sqlite/CleanedWeather.csv')
df_trips = pd.read_csv('~/data/CSV_Sqlite/Trips_Final.csv')
```

```python

df_weather['Dates']=df_weather['Dates'].apply(pd.to_datetime)
```

```python
df_trips['datetime']=df_trips['datetime'].apply(pd.to_datetime)
```

```python
df_trips
```

```python
df_trips['Dates'] = df_trips['datetime'].round('H')
```

```python
min(df_trips['JourneyTime'].unique().tolist())
```

```python
del df_trips['datetime']
```

```python
df_trips
```

```python
before_merge_trips = df_trips.sort_values('Dates')
```

```python
before_merge_weather = df_weather.sort_values('Dates')
```

```python
Final_DF =pd.merge( before_merge_trips, before_merge_weather,how='inner', on='Dates')
```

```python
Final_DF
```

```python
Final_DF.to_csv("~/data/CSV_Sqlite/WeatherTrips_Merged.csv", index=False)
```
