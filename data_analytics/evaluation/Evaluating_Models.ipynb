# Functions to evaluate and the get the best suited model per route

```python
def evaluating_models(df):
    for model in df['Model'].unique().tolist():           
        model_df = df.loc[df['Model'] == model]
        max_mae_train=model_df['MAE TRAIN'].max()
        min_mae_train= model_df['MAE TRAIN'].min()
        mean_mae_train = model_df['MAE TRAIN'].mean()
        max_mae=model_df['MAE'].max()
        min_mae= model_df['MAE'].min()
        mean_mae = model_df['MAE'].mean()

        print('Model:', model, '\nTrain Results:''\nWorst Mae:', max_mae_train, '\nBest Mae:', min_mae_train, '\nMean MAE:', mean_mae_train, '\n')
        print('Test Results:''\nWorst Mae:', max_mae, '\nBest Mae:', min_mae, '\nMean MAE:', mean_mae, '\n')


def finding_best_model_per_route(df):    
    xg_count__ = []
    rf_count__ = []
    nn_count__ = []
    routes = df['Route'].unique().tolist()
    for route in routes:
        xg__= df['MAE'].loc[(df['Route'] == route) & (df['Model'] == 'XGBoost')].tolist()[0]
        nn__ = df['MAE'].loc[(df['Route'] == route) & (df['Model'] == 'NeuralNet')].tolist()[0]
        rf__ = df['MAE'].loc[(df['Route'] == route) & (df['Model'] == 'RandomForest')].tolist()[0]
        xg__r2= df['R2'].loc[(df['Route'] == route) & (df['Model'] == 'XGBoost')].tolist()[0]
        nn__r2 = df['R2'].loc[(df['Route'] == route) & (df['Model'] == 'NeuralNet')].tolist()[0]
        rf__r2 = df['R2'].loc[(df['Route'] == route) & (df['Model'] == 'RandomForest')].tolist()[0]
        if ((xg__<=nn__ ) and (xg__r2 >=nn__r2)) or ((xg__ <=rf__) and (xg__r2>= rf__r2)):            
            xg_count__.append(route + 'XG')
        elif ((rf__ <=xg__) and (rf__r2 >= xg__r2)) or ((rf__<=nn__) and (rf__r2>=nn__r2)):            
            rf_count__.append(route+'RF')
        elif ((nn__ < xg__) and (nn__r2 >= xg__r2)) or ((nn__ < rf__) and (nn__r2 >= rf__r2)):            
            nn_count__.append(route+'NN')
    dict_ = xg_count__+rf_count__+ nn_count__
    return dict_
        
def evaluating_approaches(*args):
    approaches = ['All sub routes and outliers', 'Most Common Sub routes', 'Common Sub routes no outliers']
    for approach, df in zip(approaches,args):
        
        print('Approach: ',approach, '\n')
        print('Features:', df['Features'].unique(), '\n')
        evaluating_models(df)  
        finding_best_model_per_route(df)
        
```
